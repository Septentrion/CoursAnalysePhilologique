# Histogramme de l'espace colorimétrique
**Objectif** : visualiser la distribution des pixels dans les canaux RGB et HSV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = "../data/test/raw/voltaire_0.jpg"
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Attention: OpenCV charge en BGR ! 
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)

plt.imshow(image_rgb)
plt.title("Image originale (RGB)")
plt.axis("off")

plt.subplot(1, 3, 2)

colors = ('r', 'g', 'b')

for i, color in enumerate(colors):
    hist = cv2.calcHist([image_rgb], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)

plt.title("Histogrammes RGB")
plt.xlabel("Intensité")
plt.ylabel("Nombre de pixels")
plt.xlim([0, 256])

plt.subplot(1, 3, 3)

hist_gray = cv2.calcHist([image_gray], [0], None, [256], [0, 256])

plt.plot(hist_gray, color='black')

plt.title("Histogramme niveaux de gris")
plt.xlabel("Intensité")
plt.ylabel("Nombre de pixels")
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 4))

# Hue
plt.subplot(1, 3, 1)
hist_h = cv2.calcHist([image_hsv], [0], None, [180], [0, 180])
plt.plot(hist_h, color='red')
plt.title("Teinte (Hue)")
plt.xlabel("Angle (0-180)")
plt.ylabel("Pixels")
plt.xlim([0, 180])

# Saturation
plt.subplot(1, 3, 2)
hist_s = cv2.calcHist([image_hsv], [1], None, [256], [0, 256])
plt.plot(hist_s, color='green')
plt.title("Saturation")
plt.xlabel("Valeur")
plt.ylabel("Pixels")
plt.xlim([0, 256])

# Value
plt.subplot(1, 3, 3)
hist_v = cv2.calcHist([image_hsv], [2], None, [256], [0, 256])
plt.plot(hist_v, color='blue')
plt.title("Valeur (Brightness)")
plt.xlabel("Valeur")
plt.ylabel("Pixels")
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```
